```python
# ── People ──
first_names = [
    # Western
    "John","Sarah","Michael","Emily","David","Lisa","James","Jessica","Robert","Maria",
    "Kevin","Nina","Luke","Aaron","Tyler","Ryan","Emma","Olivia","Noah","Sophia",

    # South Asia
    "Priya","Raj","Aarav","Neha","Arjun","Isha","Rohan","Ananya","Vikram","Kavya",

    # Middle East / Muslim names
    "Omar","Zara","Fatima","Ibrahim","Ayesha","Ali","Hassan","Layla","Yusuf","Mariam",

    # East Asia
    "Chen","Mei","Yuna","Minho","Hiro","Sora","Kenji","Ling","Jin","Hana",

    # Europe / Latin
    "Elena","Sofia","Carlos","Luis","Marco","Giulia","Nico","Isabella","Mateo","Lucia"
]

last_names = [
    # Common Western
    "Smith","Johnson","Williams","Brown","Jones","Miller","Davis","Wilson","Anderson","Taylor",
    "Thomas","Moore","Jackson","White","Harris","Martin","Thompson","Walker","Hall","Young",

    # South Asian
    "Patel","Sharma","Verma","Gupta","Iyer","Reddy","Nair","Kapoor","Singh","Mehta",

    # East Asian
    "Kim","Nguyen","Chen","Wong","Lin","Zhang","Li","Park","Tran","Huang",

    # Middle Eastern / Others
    "Hassan","Ali","Khan","Rahman","Ahmed","Abdullah","Yousef","Farah",

    # European
    "Garcia","Rodriguez","Martinez","Lopez","Gonzalez","Kowalski","Novak","Silva","Costa"
]

# ── Locations ──
cities = [
    # US / Europe
    "New York","Los Angeles","Chicago","Houston","Seattle","Boston","Denver","Miami",
    "San Francisco","Austin","London","Paris","Berlin","Rome","Madrid","Amsterdam",
    "Toronto","Sydney",

    # Asia (important for realism)
    "Kathmandu","Mumbai","Delhi","Bangalore","Dhaka","Colombo","Jakarta","Manila",
    "Tokyo","Osaka","Seoul","Shanghai","Beijing","Bangkok","Singapore",

    # Middle East / Africa
    "Dubai","Abu Dhabi","Riyadh","Cairo","Nairobi","Lagos","Casablanca",

    # Latin America
    "Mexico City","Sao Paulo","Buenos Aires","Lima"
]

# ── Work ──
jobs = [
    # Tech
    "software engineer","frontend developer","backend developer","data scientist",
    "machine learning engineer","DevOps engineer","cloud architect","security analyst",

    # Business
    "product manager","project manager","business analyst","marketing manager",
    "sales executive","HR specialist","operations manager",

    # Traditional
    "doctor","nurse","teacher","lawyer","accountant","journalist","chef","designer",
    "architect","pharmacist","civil engineer","mechanical engineer",

    # Modern roles
    "AI researcher","startup founder","content creator","UX designer","data analyst"
]

companies = [
    # Big Tech
    "Google","Microsoft","Amazon","Apple","Meta","Tesla","OpenAI","Anthropic",

    # Indian / Global IT
    "Infosys","TCS","Wipro","HCL","Cognizant","Accenture","Deloitte","IBM","Oracle",

    # Startups / SaaS
    "Stripe","Shopify","Snowflake","Databricks","Zoom","Slack","Notion","Figma",
    "Salesforce","GitHub"
]

# ── Products ──
products = [
    "laptop","smartphone","wireless headphones","mechanical keyboard","4K monitor",
    "tablet","mirrorless camera","fast charger","desk lamp","gaming mouse",
    "USB-C hub","external SSD","smartwatch","printer","WiFi router","VR headset"
]

colors = ['"black"','"white"','"blue"','"red"','"silver"','"gray"','"green"','"gold"','"pink"']
colors = ['black','white','blue','red','silver','gray','green','gold','pink']

# ── Events ──
meeting_types = [
    "team standup","client call","product review","board meeting","job interview",
    "workshop","conference","training session","sprint planning","design review"
]
locations_meeting = [
    "Room A","Room B","Zoom","Google Meet","Office 3rd Floor","Conference Hall",
    "Meeting Room 1","Virtual","Teams Call","Building 2 Lobby"
]

# ── Finance ──
currencies = ['"USD"','"EUR"','"GBP"','"INR"','"NPR"','"AED"','"SGD"']
currencies = ['USD','EUR','GBP','INR','NPR','AED','SGD']
statuses = ['paid','pending','overdue','cancelled','processing','refunded']

# ── Travel ──
destinations = [
    "Paris","Tokyo","New York","London","Dubai","Singapore","Bangkok","Rome",
    "Barcelona","Bali","Kathmandu","Mumbai","Delhi","Colombo","Dhaka","Nairobi"
]
airlines = [
    "Air India","Qatar Airways","Emirates","Lufthansa","Delta","United",
    "Nepal Airlines","Turkish Airlines","Singapore Airlines","British Airways"
]
seat_classes = ['economy','business','first class']

# ── Medical ──
conditions = [
    "diabetes","hypertension","asthma","migraine","fever","fracture",
    "cold","allergy","arthritis","anemia","insomnia","depression","anxiety"
]
meds = [
    "paracetamol","ibuprofen","amoxicillin","metformin","atorvastatin",
    "cetirizine","omeprazole","aspirin","metoprolol","lisinopril"
]
dosages = ['500mg','250mg','10mg','20mg','100mg','50mg','1g','5mg','200mg']
frequencies = ['once daily','twice daily','three times a day','every 8 hours','as needed']

print(' All vocabulary defined')
print(f'   People: {len(first_names)} first × {len(last_names)} last = {len(first_names)*len(last_names):,} name combinations')
print(f'   Cities: {len(cities)}, Jobs: {len(jobs)}, Companies: {len(companies)}')
```

```python
def random_name():
    return f'{random.choice(first_names)} {random.choice(last_names)}'

def gen_person(n=500):
    """Multi-field person profiles"""
    examples = []
    for _ in range(n):
        fn = random.choice(first_names)
        ln = random.choice(last_names)
        age = random.randint(18, 65)
        city = random.choice(cities)
        job = random.choice(jobs)
        company = random.choice(companies)
        email = f'{fn.lower()}.{ln.lower()}@{company.lower().replace(" ","")}.com'
        phone = f'+1-{random.randint(200,999)}-{random.randint(100,999)}-{random.randint(1000,9999)}'

        level = random.randint(2, 6)
        if level == 2:
            text = random.choice([
                f'My name is {fn} {ln} and I am {age} years old.',
                f'{fn} {ln}, aged {age}.',
                f'Name: {fn} {ln}. Age: {age}.',
            ])
            out = {'name': f'{fn} {ln}', 'age': age}
        elif level == 3:
            text = random.choice([
                f'{fn} {ln}, {age}, lives in {city}.',
                f'I am {fn} {ln}, {age} years old from {city}.',
                f'Name: {fn} {ln}. Age: {age}. City: {city}.',
            ])
            out = {'name': f'{fn} {ln}', 'age': age, 'city': city}
        elif level == 4:
            text = random.choice([
                f"I'm {fn} {ln}, a {job} based in {city}. I'm {age} years old.",
                f'{fn} {ln} is a {age}-year-old {job} living in {city}.',
            ])
            out = {'name': f'{fn} {ln}', 'age': age, 'job': job, 'city': city}
        elif level == 5:
            text = random.choice([
                f'{fn} {ln} is a {age}-year-old {job} at {company} in {city}.',
                f'Employee: {fn} {ln}, {age}, works as {job} at {company}, based in {city}.',
            ])
            out = {'name': f'{fn} {ln}', 'age': age, 'job': job, 'company': company, 'city': city}
        else:
            text = f'{fn} {ln}, {age}, {job} at {company} in {city}. Email: {email}, Phone: {phone}.'
            out = {'name': f'{fn} {ln}', 'age': age, 'job': job, 'company': company, 'city': city, 'email': email, 'phone': phone}

        examples.append((text, out))
    return examples


def gen_orders(n=400):
    """E-commerce order records"""
    examples = []
    for _ in range(n):
        prod = random.choice(products)
        qty = random.randint(1, 10)
        price = round(random.uniform(9.99, 1999.99), 2)
        color = random.choice(colors)
        order_id = random.randint(10000, 99999)
        cust = random_name()
        total = round(qty * price, 2)

        level = random.randint(2, 4)
        if level == 2:
            text = f'Order #{order_id}: {qty} x {prod} at ${price} each. Total: ${total}.'
            out = {'order_id': order_id, 'product': prod, 'quantity': qty, 'price_per_unit': price, 'total': total}
        elif level == 3:
            text = f'{cust} ordered {qty} {color} {prod}(s) for ${price} each. Order #{order_id}.'
            out = {'customer': cust, 'product': prod, 'color': color, 'quantity': qty, 'price_per_unit': price, 'order_id': order_id}
        else:
            text = f'Customer {cust} placed order #{order_id} for {qty} units of {color} {prod} at ${price}/unit. Total: ${total}.'
            out = {'customer': cust, 'order_id': order_id, 'product': prod, 'color': color, 'quantity': qty, 'price_per_unit': price, 'total': total}

        examples.append((text, out))
    return examples


def gen_events(n=300):
    """Meeting and calendar events"""
    examples = []
    for _ in range(n):
        mtype = random.choice(meeting_types)
        loc = random.choice(locations_meeting)
        dur = random.choice([30, 45, 60, 90, 120])
        host = random_name()
        month = random.randint(1, 12)
        day = random.randint(1, 28)
        hour = random.randint(8, 17)

        level = random.randint(2, 3)
        if level == 2:
            text = f'Schedule a {mtype} on {month:02d}/{day:02d} at {hour:02d}:00 in {loc}.'
            out = {'event': mtype, 'date': f'{month:02d}/{day:02d}', 'time': f'{hour:02d}:00', 'location': loc}
        else:
            text = f'{host} is hosting a {mtype} on {month:02d}/{day:02d} at {hour:02d}:00 for {dur} minutes in {loc}.'
            out = {'host': host, 'event': mtype, 'date': f'{month:02d}/{day:02d}', 'time': f'{hour:02d}:00', 'duration_minutes': dur, 'location': loc}

        examples.append((text, out))
    return examples


def gen_financial(n=300):
    """Invoice and payment records"""
    examples = []
    for _ in range(n):
        name = random_name()
        amount = round(random.uniform(50, 50000), 2)
        currency = random.choice(currencies)
        inv_no = f'INV-{random.randint(1000,9999)}'
        status = random.choice(statuses)
        due_day = random.randint(1, 28)

        level = random.randint(2, 3)
        if level == 2:
            text = f'Invoice {inv_no} for {name}: {amount} {currency}, status: {status}.'
            out = {'invoice_id': inv_no, 'customer': name, 'amount': amount, 'currency': currency, 'status': status}
        else:
            text = f'{name} has invoice {inv_no} of {amount} {currency} due on the {due_day}th. Status: {status}.'
            out = {'customer': name, 'invoice_id': inv_no, 'amount': amount, 'currency': currency, 'due_day': due_day, 'status': status}

        examples.append((text, out))
    return examples


def gen_travel(n=300):
    """Flight and travel bookings"""
    examples = []
    for _ in range(n):
        src = random.choice(destinations)
        dst = random.choice([d for d in destinations if d != src])
        airline = random.choice(airlines)
        passengers = random.randint(1, 6)
        month = random.randint(1, 12)
        day = random.randint(1, 28)
        traveler = random_name()
        seat = random.choice(seat_classes)

        level = random.randint(2, 3)
        if level == 2:
            text = f'Book a flight from {src} to {dst} on {month:02d}/{day:02d} for {passengers} passenger(s).'
            out = {'from': src, 'to': dst, 'date': f'{month:02d}/{day:02d}', 'passengers': passengers}
        else:
            text = f'{traveler} needs {seat} tickets on {airline} from {src} to {dst} on {month:02d}/{day:02d} for {passengers} people.'
            out = {'traveler': traveler, 'class': seat, 'airline': airline, 'from': src, 'to': dst, 'date': f'{month:02d}/{day:02d}', 'passengers': passengers}

        examples.append((text, out))
    return examples


def gen_medical(n=200):
    """Patient and prescription records"""
    examples = []
    for _ in range(n):
        patient = random_name()
        age = random.randint(5, 80)
        condition = random.choice(conditions)
        med = random.choice(meds)
        dosage = random.choice(dosages)
        freq = random.choice(frequencies)

        level = random.randint(2, 3)
        if level == 2:
            text = f'Patient {patient}, age {age}, diagnosed with {condition}.'
            out = {'patient': patient, 'age': age, 'diagnosis': condition}
        else:
            text = f'{patient}, {age}, has {condition}. Prescribed {med} {dosage} {freq}.'
            out = {'patient': patient, 'age': age, 'condition': condition, 'medication': med, 'dosage': dosage, 'frequency': freq}

        examples.append((text, out))
    return examples


def gen_refusals():
    """Ambiguous inputs → empty JSON {}"""
    ambiguous = [
        'Hello, how are you?', 'What is the capital of France?', 'Tell me a joke.',
        'What is 2 + 2?', 'Can you help me?', 'The weather is nice today.',
        'I enjoy reading books.', 'Nice to meet you.', 'Good morning!',
        'Please explain machine learning.', 'What time is it?', 'I like pizza.',
        'How does the internet work?', 'The sky is blue.', 'Tell me something interesting.',
        'Is it going to rain tomorrow?', 'Recommend a movie.', 'What is AI?',
        'Translate this to French.', 'Write me a poem.', 'Calculate 100 * 3.',
    ]
    return [(q, {}) for q in ambiguous]


print(' All generator functions defined')
print('Categories: Person | Orders | Events | Financial | Travel | Medical | Refusals')
```

```python
print('Generating data...')

all_raw = []
all_raw += gen_person(500)
print(f'   Person profiles: {len(all_raw)}')

prev = len(all_raw)
all_raw += gen_orders(400)
print(f'   Orders: +{len(all_raw)-prev}')

prev = len(all_raw)
all_raw += gen_events(300)
print(f'   Events: +{len(all_raw)-prev}')

prev = len(all_raw)
all_raw += gen_financial(300)
print(f'   Financial: +{len(all_raw)-prev}')

prev = len(all_raw)
all_raw += gen_travel(300)
print(f'   Travel: +{len(all_raw)-prev}')

prev = len(all_raw)
all_raw += gen_medical(200)
print(f'   Medical: +{len(all_raw)-prev}')

prev = len(all_raw)
all_raw += gen_refusals()
print(f'   Refusals: +{len(all_raw)-prev}')

print(f'\n📊 Total raw examples: {len(all_raw)}')
```

```python
def format_example(input_text, output_dict):
    instruction = random.choice(INSTRUCTIONS)
    output_str = json.dumps(output_dict)
    return {
        'instruction': instruction,
        'input': input_text,
        'output': output_str,
        'text': f'### Instruction:\n{instruction}\n\n### Input:\n{input_text}\n\n### Response:\n{output_str}'
    }

formatted = [format_example(inp, out) for inp, out in all_raw]

# Shuffle
random.shuffle(formatted)

print(f' Formatted {len(formatted)} examples')
print('\nSample example:')
sample = formatted[0]
print(f'  INSTRUCTION: {sample["instruction"]}')
print(f'  INPUT: {sample["input"]}')
print(f'  OUTPUT: {sample["output"]}')
```

```python
valid = []
invalid_count = 0
invalid_examples = []

for ex in formatted:
    try:
        # Check 1: Input not empty
        assert ex['input'].strip(), 'Empty input'

        # Check 2: Output is valid JSON
        parsed_out = json.loads(ex['output'])

        # Check 3: Output is a dict (not list, not string)
        assert isinstance(parsed_out, dict), f'Output is not a dict: {type(parsed_out)}'

        # Check 4: All keys are strings
        for k in parsed_out.keys():
            assert isinstance(k, str), f'Key is not string: {k}'

        # Check 5: text field contains both instruction and input
        assert '### Instruction:' in ex['text'], 'Missing instruction header'
        assert '### Input:' in ex['text'], 'Missing input header'
        assert '### Response:' in ex['text'], 'Missing response header'

        valid.append(ex)

    except Exception as e:
        invalid_count += 1
        invalid_examples.append((ex.get('input', '?'), str(e)))

print(f' Valid examples: {len(valid)}')
print(f'❌ Invalid examples removed: {invalid_count}')

if invalid_examples:
    print('\nInvalid examples:')
    for inp, err in invalid_examples[:5]:
        print(f'  INPUT: {inp[:60]}... | ERROR: {err}')

# Check for duplicates
inputs = [ex['input'] for ex in valid]
unique_inputs = set(inputs)
print(f'\n📊 Unique inputs: {len(unique_inputs)} / {len(valid)}')
if len(unique_inputs) < len(valid):
    print(f'  ⚠️  {len(valid) - len(unique_inputs)} duplicates found — removing...')
    seen = set()
    deduped = []
    for ex in valid:
        if ex['input'] not in seen:
            seen.add(ex['input'])
            deduped.append(ex)
    valid = deduped
    print(f'   After dedup: {len(valid)} examples')
```

```python
# Field frequency analysis
from collections import Counter

field_counter = Counter()
output_lengths = []
empty_outputs = 0

for ex in valid:
    out = json.loads(ex['output'])
    if not out:
        empty_outputs += 1
    for k in out.keys():
        field_counter[k] += 1
    output_lengths.append(len(out))

print('📊 DATASET STATISTICS')
print('=' * 50)
print(f'Total examples:      {len(valid)}')
print(f'Refusal examples:    {empty_outputs} (output = {{}})')
print(f'Non-empty outputs:   {len(valid) - empty_outputs}')
print(f'Avg fields/output:   {sum(output_lengths)/len(output_lengths):.2f}')
print(f'Max fields:          {max(output_lengths)}')
print(f'Min fields (non-0):  {min(x for x in output_lengths if x > 0)}')
print()
print('Top 15 most common output fields:')
for field, count in field_counter.most_common(15):
    bar = '█' * (count // 10)
    print(f'  {field:<20} {count:>5}  {bar}')

print()
print(' Dataset looks healthy if:')
print('   - Multiple fields appear frequently (not just "name")')
print('   - Avg fields/output > 2.5')
print('   - Refusal examples are < 5% of total')
```

```python
# Split
n = len(valid)
train_end = int(n * 0.80)
val_end   = int(n * 0.90)

train_data = valid[:train_end]
val_data   = valid[train_end:val_end]
test_data  = valid[val_end:]

print(f'📊 Split summary:')
print(f'  Train: {len(train_data)} examples ({len(train_data)/n*100:.1f}%)')
print(f'  Val:   {len(val_data)} examples ({len(val_data)/n*100:.1f}%)')
print(f'  Test:  {len(test_data)} examples ({len(test_data)/n*100:.1f}%)')

# Save
def save_jsonl(name, data):
    path = os.path.join(OUT_DIR, f'{name}.jsonl')
    with open(path, 'w', encoding='utf-8') as f:
        for row in data:
            f.write(json.dumps(row) + '\n')
    size_kb = os.path.getsize(path) / 1024
    print(f'   Saved {name}.jsonl — {len(data)} examples — {size_kb:.1f} KB')

print('\nSaving files...')
save_jsonl('train', train_data)
save_jsonl('val', val_data)
save_jsonl('test', test_data)

print(f'\n All files saved to: {os.path.abspath(OUT_DIR)}')
print('📁 Files created:')
for f in ['train.jsonl', 'val.jsonl', 'test.jsonl']:
    fpath = os.path.join(OUT_DIR, f)
    if os.path.exists(fpath):
        print(f'   {f}')
```

```python
# Reload and verify
def load_jsonl(name):
    path = os.path.join(OUT_DIR, f'{name}.jsonl')
    rows = []
    with open(path, encoding='utf-8') as f:
        for line in f:
            rows.append(json.loads(line.strip()))
    return rows

train_check = load_jsonl('train')
val_check   = load_jsonl('val')
test_check  = load_jsonl('test')

print(' Files reloaded successfully')
print(f'  train.jsonl: {len(train_check)} rows')
print(f'  val.jsonl:   {len(val_check)} rows')
print(f'  test.jsonl:  {len(test_check)} rows')

# Verify all outputs are valid JSON
errors = 0
for split_name, split in [('train', train_check), ('val', val_check), ('test', test_check)]:
    for ex in split:
        try:
            json.loads(ex['output'])
        except:
            errors += 1
            print(f'  ❌ Invalid JSON in {split_name}: {ex["output"][:80]}')

if errors == 0:
    print('\n ALL outputs are valid JSON — dataset is ready!')
else:
    print(f'\n❌ {errors} invalid JSON outputs found — rerun the validation cell!')

print('\n5 sample training examples:')
print('-' * 60)
for ex in train_check[:5]:
    print(f'INPUT:  {ex["input"]}')
    print(f'OUTPUT: {ex["output"]}')
    print()
```
